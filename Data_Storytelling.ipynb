```python
# importing necessary library  
import pandas as pd
temporal_data = pd.read_csv('temporal.csv')
temporal_data.head()
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
# Making sure month column has a datetime format
format_dict = {'data science':'${0:,.2f}', 'Mes':'{:%m-%Y}', 'machine learning':'{:.2%}'}
temporal_data['Mes'] = pd.to_datetime(temporal_data['Mes'])
#Applying style to visualization
temporal_data.head().style.format(format_dict)
```

```python
format_dict = {'Mes':'{:%m-%Y}'}
# simplifying format dic with values that makes sense for the dataset
temporal_data.head().style.format(format_dict).highlight_max(color='darkgreen').highlight_min(color='#ff0000')
```

```python
temporal_data.head(10).style.format(format_dict).background_gradient(subset=['data science', 'machine learning'], cmap='BuGn')
```

```python
# Displaying the data values using bars
temporal_data.head().style.format(format_dict).bar(color='red', subset=['data science', 'deep learning'])
```

```python
# combining the above functions to generate a more complex visualization as seen below

temporal_data.head(10).style.format(format_dict).background_gradient(subset=['data science', 'machine learning'],
                                                                    cmap='BuGn').highlight_max(color='orange')
```

```python
plt.plot(temporal_data['Mes'], temporal_data['data science'], label='data science')
plt.plot(temporal_data['Mes'], temporal_data['machine learning'], label='machine learning')
plt.plot(temporal_data['Mes'], temporal_data['deep learning'], label='deep learning')
plt.xlabel('Date')
plt.ylabel('Popularity')
plt.title('Popularity of AI terms by date')
plt.grid(True)
plt.legend();
```

```python
# Importing seaborn library
import seaborn as sns
sns.set()
sns.scatterplot(temporal_data['Mes'], temporal_data['data science']);
```

We can add information of more than two variables in the same graph. For this we will use colors and sizes. We also make a different graph according to the value of the category column:

```python
sns.relplot(x='Mes', y='deep learning', hue='data science', size='machine learning', col='categorical',
           data=temporal_data);
```

Using heatmap to show all the correlations between variables in the dataset

```python
sns.heatmap(temporal_data.corr(), annot=True, fmt='.2f');
```

```python
sns.pairplot(temporal_data);
```

using pairplot in showing the charts segmented according to the values of the categorical variable.

```python
sns.pairplot(temporal_data, hue='categorical');
```

Using the ViolinPlot to visualize the data

```python
sns.catplot(x='categorical', y='data science', kind='violin', data=temporal_data);
```
